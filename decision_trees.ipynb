```python
import pandas as pd
dados = pd.read_csv('C:/Users/Andre/Documents/DidaticaTech/vinho.csv')
dados.head()
```

```python
from sklearn.preprocessing import LabelEncoder
pi = LabelEncoder()
dados['style'] = pi.fit_transform(dados['style'])
```

```python
x = dados.drop('style', axis=1)
y = dados['style']
import numpy as np
X = x.values 
Y = y.values.ravel() 
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold as skd
from sklearn.tree import DecisionTreeClassifier as dtc
skfold = skd(n_splits=5)
modelo = dtc()
resultado = cross_val_score(modelo,X,Y,cv=skfold)
print('auracia:', resultado.mean())
```

```python
nomes = ['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar', 'chlorides', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'density', 'pH', 'sulphates','quality',	'style']
```

```python
import graphviz
from sklearn.tree  import export_graphviz
```

```python
arquivo = 'C:/Users/Andre/Documents/DidaticaTech/graphviz1.csv'
modelo.fit(X,Y)
export_graphviz(modelo, out_file=arquivo, feature_names = nomes)
with open(arquivo) as aberto:
    grafico_dot = aberto.read()
h = graphviz.Source(grafico_dot)
h.view()
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import GridSearchCV 
min_split = ([2,5,7,9,10,20,30,40,50,60,45])
max_level = ([3,5,9,10,14,30,34,19,11,17,9,20,])
algoritmo = ['gini','entropy']
valores_grid = {'min_samples_split':min_split,'max_depth':max_level,'criterion':algoritmo}
modelo = DecisionTreeClassifier()
gdt = GridSearchCV(estimator=modelo,param_grid=valores_grid,cv=5)
gdt.fit(X,Y)
print('acuracia',gdt.best_score_)
print("Melhores parâmetros:", gdt.best_params_)
```
